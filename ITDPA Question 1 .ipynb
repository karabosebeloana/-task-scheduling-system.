```python
import heapq
from datetime import datetime
import os

class Task:
    """
    Task class to store task information including description, due date, 
    priority level, and estimated time to complete.
    """
    def __init__(self, description, due_date, priority, estimated_time):
        """
        Initialize a Task object.
        
        Args:
            description (str): Description of the task
            due_date (str): Due date in format DD/MM/YYYY
            priority (int): Priority level (lower number = higher priority)
            estimated_time (float): Estimated time to complete in hours
        """
        self.description = description
        self.due_date = due_date
        self.priority = priority
        self.estimated_time = estimated_time
    
    def print_task(self):
        """Print the task's information in a formatted manner."""
        print(f"\n{'='*60}")
        print(f"Task Description: {self.description}")
        print(f"Due Date: {self.due_date}")
        print(f"Priority Level: {self.priority}")
        print(f"Estimated Time: {self.estimated_time} hours")
        print(f"{'='*60}")
    
    def __lt__(self, other):
        """
        Less than comparison for heap ordering.
        Lower priority number means higher priority.
        """
        return self.priority < other.priority
    
    def __str__(self):
        """String representation of the task."""
        return f"{self.description} | Due: {self.due_date} | Priority: {self.priority} | Time: {self.estimated_time}h"


class TaskScheduler:
    """
    Scheduler class to manage tasks using a priority queue (min-heap).
    Tasks with lower priority numbers are executed first.
    """
    def __init__(self):
        """Initialize the scheduler with an empty priority queue."""
        self.task_heap = []
        self.task_history = []  # Keep track of all tasks for file operations
    
    def add_task(self, task):
        """
        Add a task to the scheduler.
        
        Args:
            task (Task): Task object to be added
        """
        heapq.heappush(self.task_heap, task)
        self.task_history.append(task)
        print(f"\n✓ Task '{task.description}' added successfully with priority {task.priority}!")
    
    def get_next_task(self):
        """
        Retrieve and remove the highest priority task from the scheduler.
        
        Returns:
            Task: The highest priority task, or None if scheduler is empty
        """
        if self.task_heap:
            task = heapq.heappop(self.task_heap)
            print(f"\n→ Retrieved highest priority task:")
            task.print_task()
            return task
        else:
            print("\n! No tasks in the scheduler.")
            return None
    
    def print_task_list(self):
        """Print all tasks currently in the scheduler ordered by priority."""
        if not self.task_heap:
            print("\n! The task scheduler is empty.")
            return
        
        print(f"\n{'='*60}")
        print(f"{'CURRENT TASK LIST (Ordered by Priority)':^60}")
        print(f"{'='*60}")
        
        # Create a sorted copy to display without modifying the heap
        sorted_tasks = sorted(self.task_heap, key=lambda x: x.priority)
        
        for idx, task in enumerate(sorted_tasks, 1):
            print(f"\n{idx}. {task}")
        
        print(f"\n{'='*60}")
        print(f"Total tasks: {len(self.task_heap)}")
        print(f"{'='*60}")
    
    def save_tasks_to_file(self, filename="tasks.txt"):
        """
        Save all task details to a file in order of occurrence (task history).
        
        Args:
            filename (str): Name of the file to save tasks to
        """
        try:
            with open(filename, 'w') as file:
                file.write("TASK SCHEDULING SYSTEM - TASK HISTORY\n")
                file.write("=" * 60 + "\n")
                file.write(f"Generated on: {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}\n")
                file.write("=" * 60 + "\n\n")
                
                if not self.task_history:
                    file.write("No tasks recorded.\n")
                else:
                    for idx, task in enumerate(self.task_history, 1):
                        file.write(f"\nTask {idx}:\n")
                        file.write(f"  Description: {task.description}\n")
                        file.write(f"  Due Date: {task.due_date}\n")
                        file.write(f"  Priority: {task.priority}\n")
                        file.write(f"  Estimated Time: {task.estimated_time} hours\n")
                        file.write("-" * 60 + "\n")
            
            print(f"\n✓ Tasks successfully saved to '{filename}'!")
        except Exception as e:
            print(f"\n✗ Error saving tasks to file: {e}")
    
    def read_tasks_from_file(self, filename="tasks.txt"):
        """
        Read and display task list from a file.
        
        Args:
            filename (str): Name of the file to read tasks from
        """
        try:
            if not os.path.exists(filename):
                print(f"\n✗ File '{filename}' not found.")
                return
            
            with open(filename, 'r') as file:
                content = file.read()
                print(f"\n{'='*60}")
                print("TASKS FROM FILE")
                print(f"{'='*60}")
                print(content)
                print(f"{'='*60}")
        except Exception as e:
            print(f"\n✗ Error reading tasks from file: {e}")
    
    def is_empty(self):
        """Check if the scheduler is empty."""
        return len(self.task_heap) == 0


def display_menu():
    """Display the main menu options."""
    print(f"\n{'='*60}")
    print(f"{'TASK SCHEDULING SYSTEM - MAIN MENU':^60}")
    print(f"{'='*60}")
    print("1. Add a new task")
    print("2. Get next task (highest priority)")
    print("3. View all tasks")
    print("4. Save tasks to file")
    print("5. Read tasks from file")
    print("6. Exit")
    print(f"{'='*60}")


def get_task_input():
    """
    Get task details from user input.
    
    Returns:
        Task: A new Task object with user-provided details
    """
    print(f"\n{'='*60}")
    print("ENTER TASK DETAILS")
    print(f"{'='*60}")
    
    description = input("Task Description: ").strip()
    
    while True:
        due_date = input("Due Date (DD/MM/YYYY): ").strip()
        try:
            # Validate date format
            datetime.strptime(due_date, '%d/%m/%Y')
            break
        except ValueError:
            print("✗ Invalid date format. Please use DD/MM/YYYY.")
    
    while True:
        try:
            priority = int(input("Priority Level (1=highest, lower number=higher priority): "))
            if priority > 0:
                break
            else:
                print("✗ Priority must be a positive number.")
        except ValueError:
            print("✗ Please enter a valid number.")
    
    while True:
        try:
            estimated_time = float(input("Estimated Time (hours): "))
            if estimated_time > 0:
                break
            else:
                print("✗ Time must be greater than 0.")
        except ValueError:
            print("✗ Please enter a valid number.")
    
    return Task(description, due_date, priority, estimated_time)


def main():
    """Main function to run the task scheduling application."""
    scheduler = TaskScheduler()
    
    print(f"\n{'='*60}")
    print(f"{'WELCOME TO TASK SCHEDULING SYSTEM':^60}")
    print(f"{'='*60}")
    print("\nThis application helps you manage your academic tasks")
    print("by priority, ensuring the most important tasks are")
    print("completed first!")
    
    while True:
        display_menu()
        
        choice = input("\nEnter your choice (1-6): ").strip()
        
        if choice == '1':
            # Add a new task
            task = get_task_input()
            scheduler.add_task(task)
        
        elif choice == '2':
            # Get next task
            scheduler.get_next_task()
        
        elif choice == '3':
            # View all tasks
            scheduler.print_task_list()
        
        elif choice == '4':
            # Save tasks to file
            filename = input("\nEnter filename (press Enter for 'tasks.txt'): ").strip()
            if not filename:
                filename = "tasks.txt"
            scheduler.save_tasks_to_file(filename)
        
        elif choice == '5':
            # Read tasks from file
            filename = input("\nEnter filename (press Enter for 'tasks.txt'): ").strip()
            if not filename:
                filename = "tasks.txt"
            scheduler.read_tasks_from_file(filename)
        
        elif choice == '6':
            # Exit application
            print(f"\n{'='*60}")
            print("Thank you for using Task Scheduling System!")
            print("Stay productive! ")
            print(f"{'='*60}\n")
            break
        
        else:
            print("\n✗ Invalid choice. Please enter a number between 1 and 6.")
        
        # Pause before showing menu again
        input("\nPress Enter to continue...")


if __name__ == "__main__":
    main()
```
